## 2.Windows functions using python.
#### Dataset : “storesales.csv”
#### Problem statement:-

    1).	Replicate the functionality of SQL ROW_NUMBER() windows function
    2).	Replicate the functionality of SQL RANK() windows function
    3).  Replicate the functionality of SQL DENSE_RANK() windows function
    4).	Replicate the functionality of SQL LAG() windows function
    5).	Replicate the functionality of SQL LEAD() windows function

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
```

```python
# importing the dataset as a dataframe
df=pd.read_csv("storesales.csv")
df.head()
```

```python
## 1st_approach :- 
class windows_func():
    def __init__(self,df,partition_colm,order_by_colm,order_type):
        self.df = df
        self.partition_colm=partition_colm
        self.order_by_colm=order_by_colm
        self.order_type=order_type
        
    def row_num_deep_code(self):
        sorted_df=self.df.sort_values(by=[self.partition_colm,self.order_by_colm],ascending=self.order_type) 
        sorted_df["row_num"]=np.NAN
        for j in list(sorted_df.store.unique()):
            sorted_df.row_num[(sorted_df.store==j)]=[*range(1,len([k for k in sorted_df.store if k==j])+1)]
        return sorted_df
    
    def row_num(self):
        self.df['Row Num'] = self.df.groupby([self.partition_colm])[self.order_by_colm].rank(method='first') ## gives rank based on ascending values here
        self.df.sort_values(by= [self.partition_colm,self.order_by_colm], inplace = True)
        fdf = self.df.reset_index(drop = True)
        return fdf
    
    def rank(self):
        self.df['Rank'] = self.df.groupby([self.partition_colm])[self.order_by_colm].rank(method='min')## gives rank based on ascending values here
        self.df.sort_values(by= [self.partition_colm,self.order_by_colm], inplace = True)
        fdf = self.df.reset_index(drop = True)
        return fdf
    
    def dense_rank(self):
        self.df['Dense Rank'] = self.df.groupby([self.partition_colm])[self.order_by_colm].rank(method='dense')## gives rank based on ascending values here
        self.df.sort_values(by= [self.partition_colm,self.order_by_colm], inplace = True)
        fdf = self.df.reset_index(drop = True)
        return fdf   
    
    def rank_lag(self):
        self.df.sort_values(by= [self.partition_colm,self.order_by_colm], inplace = True)
        self.df['Lag'] = self.df.groupby([self.partition_colm])[self.order_by_colm].shift(1)## gives rank based on ascending values here
        fdf = self.df.reset_index(drop = True)
        return fdf   
    
    def rank_lead(self):
        self.df.sort_values(by= [self.partition_colm,self.order_by_colm], inplace = True)
        self.df['Lead'] = self.df.groupby([self.partition_colm])[self.order_by_colm].shift(-1)## gives rank based on ascending values here
        fdf = self.df.reset_index(drop = True)
        return fdf
```

####    1).	Replicate the functionality of SQL ROW_NUMBER() windows function
        a).method name (using user defined method): row_num_deep_code()
        b).method name (using built in method): row_num()
           .rank(method='first')
####     2).	Replicate the functionality of SQL RANK() windows function
        method name (using built in method): rank()
        .rank(method='min')
####     3).  Replicate the functionality of SQL DENSE_RANK() windows function
        method name (using built in method): dense_rank()
        .rank(method='dense')
####     4).	Replicate the functionality of SQL LAG() windows function
        method name (using built in method): rank_lag()
        .shift(1)
####     5).	Replicate the functionality of SQL LEAD() windows function
        method name (using built in method): rank_lead()
        .shift(-1)

```python
class Inventory_Management(): 
    """Inventory dataset"""
    def __init__(self,furn_sub_iq,spo_sub_iq,food_sub_iq, furn_sub_oq,spo_sub_oq,food_sub_oq):
        self.furn_sub_iq=furn_sub_iq
        self.spo_sub_iq=spo_sub_iq
        self.food_sub_iq=food_sub_iq
        self.furn_sub_oq=furn_sub_oq
        self.spo_sub_oq=spo_sub_oq
        self.food_sub_oq=food_sub_oq

    def available_inventory(self):
        """available items and quantity details"""
        print("===================================================================================================================")
        print("                                  -- Avialable Inventory --                      ","Date:-",datetime.now())
        print("===================================================================================================================")
        print("%s %10s %10s %10s %1s %10s"% ("Sr.No","Category" ,"Sub_category","Unit_price($)", "Initial_quantity","TOTAL($)"))
        print("-------------------------------------------------------------------------------------------------------------------") 
        for i,p in enumerate(Category):
            if i==0:
                furn_total_sales_list=[]
                furn_total_sales_avlb = []
                for j,k,l,m in zip(Furn_sub,furn_sub_up,self.furn_sub_iq,self.furn_sub_oq):
                    print("%2d %13s %10s %10.2f %15d %18.2f"  %(i+1,p,j,k,l-m,k*(l-m)))
                    furn_total_sales_list.append(k*(l-m))
                    furn_total_sales_avlb.append((l-m))
                    i=i+1
                print()
            elif i==1:
                spo_total_sales_list=[]
                spo_total_sales_avlb = []
                for j,k,l,m in zip(Spo_sub,spo_up,self.spo_sub_iq,self.spo_sub_oq):
                    print("%2d %13s %10s %10.2f %15d %18.2f"  %(i,p,j,k,l-m,k*(l-m)))
                    spo_total_sales_list.append(k*(l-m))
                    spo_total_sales_avlb.append((l-m))
                    i=i+1
                print()
            else :
                food_total_sales_list=[]
                food_total_sales_avlb = []
                for j,k,l,m in zip(Food_sub,food_sub_up,self.food_sub_iq,self.food_sub_oq):
                    print("%2d %13s %10s %10.2f %15d %18.2f"  %(i-1,p,j,k,l-m,k*(l-m)))
                    food_total_sales_avlb.append((l-m))
                    food_total_sales_list.append(k*(l-m))
                    i=i+1
                print()
        print("---------------------------------------------------------------------------------------------------------------")
        print("* Furniture total sales =",sum(furn_total_sales_list),"$")
        print("* Sports total sales =",sum(spo_total_sales_list),"$")
        print("* Food total sales =",sum(food_total_sales_list),"$")
        print()
        print("==> Total_sales =",sum([sum(furn_total_sales_list),sum(spo_total_sales_list),sum(food_total_sales_list)]),"$")
        print("================================================================================================================")
        
        return furn_total_sales_avlb,spo_total_sales_avlb,food_total_sales_avlb


    def ordered_inventory(self):
        cust_name=input("Name of Customer : ")
        Adress=input("Enter adress : ")
        phone_no=int(input("Phone Number : "))
        #check which categories
        order_type=input("Do you want to purchase Farniture Items : ")
        if order_type == "yes" or order_type == "Yes":
            fur_oq_chairs=int(input("enter number of chairs : "))
            fur_oq_tables=int(input("enter number of tables : "))
            fur_oq_wardrobe=int(input("enter number of wardrobe : "))
            fur_oq_beds=int(input("enter number of beds : "))
            fur_oq_couch=int(input("enter number of couches : "))
            fur_oq_antique=int(input("enter number of antique : "))
            fur_oq_storage=int(input("enter number of storage : "))
            furn_sub_oq=[fur_oq_chairs,fur_oq_tables,fur_oq_wardrobe,fur_oq_beds,fur_oq_couch,fur_oq_antique,fur_oq_storage]
        else:
            furn_sub_oq = [0,0,0,0,0,0,0]
        
        order_type=input("Do you want to purchase sport items: ")
        if order_type =='yes' or order_type == "Yes":
            print(furn_sub_oq)
            spo_oq_shoes=int(input("enter number of shoes : "))
            spo_oq_bags=int(input("enter number of bags"))
            spo_oq_backpack=int(input("enter number of backpack : "))
            spo_oq_bottles=int(input("enter number of bottles : "))
            spo_oq_towels=int(input("enter number of towels : "))
            spo_sub_oq=[spo_oq_shoes,spo_oq_bags,spo_oq_backpack,spo_oq_bottles,spo_oq_towels]
        else:
            spo_sub_oq = [0,0,0,0,0]
            
        order_type=input("Do you want to purchase Food items: ")
        if order_type =='yes' or order_type == "Yes":
            food_oq_chips=int(input("enter number of chips packetes : "))
            food_oq_bread=int(input("enter number of bread packetes : "))
            food_oq_chocolate=int(input("enter number of chocolate : "))
            food_oq_biscuits=int(input("enter number of biscuits : "))
            food_oq_rice=int(input("enter kg of rice : "))
            food_oq_wheat=int(input("enter kg of wheat : "))
            food_oq_sugar=int(input("enter kg of sugar : "))
            food_oq_flour=int(input("enter kg of flour : "))
            food_sub_oq=[food_oq_chips,food_oq_bread,food_oq_chocolate,food_oq_biscuits,food_oq_rice,food_oq_wheat,food_oq_sugar,food_oq_flour]
        else:
            food_sub_oq = [0,0,0,0,0,0,0,0]
        
        
        print()
        print()
        print("===================================================================================================================")
        print("                                  -- Ordered Inventory --                      ","Date:-",datetime.now())
        print("===================================================================================================================")
        print()
        print("Customer_name :",cust_name)
        print("Adress :",Adress)
        print("phone_no :",phone_no)
        print()
        print("-------------------------------------------------------------------------------------------------------------------")
        print("%s %10s %10s %10s %1s %10s"% ("Sr.No","Category" ,"Sub_category","Unit_price($)", "Ordered_quantity","TOTAL($)"))
        print("-------------------------------------------------------------------------------------------------------------------")
        
        for i,p in enumerate(Category):
            if i==0:
                furn_total_sales_list=[]
                for j,k,l in zip(Furn_sub,furn_sub_up,furn_sub_oq):  
                    if l!=0:
                        if l > furn_sub_iq[i]:
                            print("max u can order= ",furn_sub_iq[i])
                            print("%2d %13s %10s %10.2f %15d %18.2f"  %(i+1,p,j,k,furn_sub_iq[i],k*furn_sub_iq[i]))
                            furn_total_sales_list.append(k*furn_sub_iq[i])
                        else:
                            print("%2d %13s %10s %10.2f %15d %18.2f"  %(i+1,p,j,k,l,k*l))   ## L CHange honar yamdhe fakt
                            furn_total_sales_list.append(k*l)
                        i=i+1
                print()
            elif i==1:
                Spo_total_sales_list=[]
                for j,k,l in zip(Spo_sub,spo_up,spo_sub_oq):
                    if l!=0:
                        if l > spo_sub_iq[i-1]:
                            print("max u can order= ",furn_sub_iq[i-1])
                            print("%2d %13s %10s %10.2f %15d %18.2f"  %(i,p,j,k,furn_sub_iq[i],k*spo_sub_iq[i-1]))
                            Spo_total_sales_list.append(k*spo_sub_iq[i-1])
                        else:
                            print("%2d %13s %10s %10.2f %15d %18.2f"  %(i,p,j,k,l,k*l))
                            Spo_total_sales_list.append(k*l)
                        i=i+1
                print()
            else :
                food_total_sales_list=[]
                for j,k,l in zip(Food_sub,food_sub_up,food_sub_oq):
                    if l!=0:
                        if l > food_sub_iq[i-2]:
                            print("max u can order= ",furn_sub_iq[i-2])
                            print("%2d %13s %10s %10.2f %15d %18.2f"  %(i-1,p,j,k,furn_sub_iq[i],k*food_sub_iq[i-2]))
                            food_total_sales_list.append(k*food_sub_iq[i-2])
                        else:
                            print("%2d %13s %10s %10.2f %15d %18.2f"  %(i-1,p,j,k,l,k*l))
                            food_total_sales_list.append(k*l)
                        i=i+1
                print()
        print("---------------------------------------------------------------------------------------------------------------")
        print("* Furniture total =",sum(furn_total_sales_list),"$")
        print("* Sports total =",sum(Spo_total_sales_list),"$")
        print("* Food total =",sum(food_total_sales_list),"$")
        print()
        print("==> Total_bill_amount =",sum([sum(furn_total_sales_list),sum(Spo_total_sales_list),sum(food_total_sales_list)]),"$")
        print("================================================================================================================")

        return furn_sub_oq,spo_sub_oq,food_sub_oq
```
